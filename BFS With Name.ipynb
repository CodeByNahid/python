```python

from collections import deque
def bfs(graph, start):
    queue=deque([start])
    visited=set([start])
    while queue:
        current_node=queue.popleft()
        for children in graph[current_node]:
            if children not in visited:
                queue.append(children)
                visited.add(children)
        if queue:
            print(current_node,end="->")
        else:
            print(current_node,end="")
            
```

```python
graph={
    "A": ["B","C"],
    "B": ["D","E"],
    "C": ["F"],
    "D": [],
    "E": ["F"],
    "F": []
}
bfs(graph,"A")


```
