```python
import sklearn as sk
import matplotlib as mtp
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris 
import pandas as pd
```

```python
iris_data=load_iris()
print(iris_data)
```

```python
iris_df=pd.DataFrame(data=iris_data.data,columns=iris_data["feature_names"])
print(iris_df)
```

```python
iris_df["species"]=iris_data.target
print(iris_df)
```

```python
iris_df["species"] = [iris_data.target_names[i] for i in iris_data.target]
```

```python
sns.lmplot(data=iris_df, x='sepal length (cm)', y='sepal width (cm)', hue='species')
plt.show()
```

```python
sns.barplot(data=iris_df, x='species', y='sepal length (cm)')
plt.show()
```

```python
sns.displot(data=iris_df, x='sepal length (cm)', hue='species', kind='kde')
plt.show()
```

```python
sns.heatmap(iris_df.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
sns.swarmplot(data=iris_df, x='species', y='sepal length (cm)')
plt.show()
```

```python
sns.catplot(data=iris_df, x='species', y='sepal length (cm)', kind='box')
plt.show()
```

```python
sns.histplot(data=iris_df, x='sepal length (cm)', hue='species', element='step')
plt.show()
```

```python
sns.violinplot(data=iris_df, x='species', y='sepal length (cm)')
plt.show()
```
