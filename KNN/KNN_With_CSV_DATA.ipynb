```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter
```

```python
data=pd.read_csv('data.csv')
data.head()
```

```python
def KNN(data, new_sample, k):
    dis = []
    for _, ani in data.iterrows():
        ecd = np.linalg.norm(np.array([ani['Height(cm)'], ani['Weight(kg)']]) - np.array(new_sample))
        dis.append([ecd, ani['Animal']])
    dis = sorted(dis)[:k]
    votes = [vote[1] for vote in dis]
    result = Counter(votes).most_common(1)[0][0]
    print("The Animal is a:", result)
    sns.scatterplot(data=data, x="Weight(kg)", y="Height(cm)", hue="Animal", s=100)
    plt.scatter(x=new_sample[1], y=new_sample[0], marker="*", s=200, c='black', label="New Sample")
    plt.legend()
    plt.title("KNN Classification")
    plt.show()
```

```python

h = int(input("Enter the value of Height (cm): "))
w = int(input("Enter the value of Weight (kg): "))

# Run KNN
KNN(data, new_sample=[h, w], k=5)
```
