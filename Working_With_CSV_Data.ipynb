```python
def KNN(xt,yt,xTest,k):
    y_pre=[]
    for s in xTest:
        distances=[np.linalg.norm(np.array(s)-np.array(xts)) for xts in xt]
        v=np.argsort(distances)[:k]
        vc=[yt[i] for i in v]
        vr=Counter(vc).most_common()[0][0]
        y_pre.append(vr)
    return np.array(y_pre)
        
```
